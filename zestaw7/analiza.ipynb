# Data Analysis
This notebook presents bar plot with counted number of posts from each year from www.dogmatykarnisty.pl

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dogmaty.csv')
df.head()
```

```python
papers_per_year = df.groupby('year').size()
plt.barh(papers_per_year.index, papers_per_year)
plt.title("Wpisy na portalu Dogmaty Karnisty w poszczególnych latach")
plt.xlabel("Liczba wpisów")
plt.ylabel("Rok")
for year, value in papers_per_year.items():
    plt.text(value + 1, year, str(value), va='center')
plt.show()
```
